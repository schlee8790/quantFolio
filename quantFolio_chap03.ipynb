## 3.5 결측치 처리하기

```python
df.head()
```

```python
df.info()
```

```python
df.head().isnull()
```

```python
# 결측치 삭제하기
df.dropna()
```

```python
# subset 지정한 열에 결측치가 있는 경우 해당 행을 삭제하기 (axis=0 은 row 방향)
df.dropna(subset=['age'], axis=0)
```

```python
# 결측지가 있는 열 삭제하기
df.dropna(axis=1)
```

```python
# 결측치가 300개 이상 있는 열을 삭제하기
df.dropna(axis=1, thresh=300)
```

```python
# 결측치 대체하기
df_2 = df.copy()

df_2.head(6)
```

```python
mean_age = df_2['age'].mean()
print(mean_age)
```

```python
# age 결측치를 평균으로 대체하기 (inplace=True: 원본을 변경하는 경우)
# df_2['age'].fillna(mean_age, inplace=True)
df_2.fillna({'age': mean_age}, inplace=True)
df_2['age'].head(6)
```

```python
# 결측치를 문자로 변경하는 경우
df_2.fillna({'embark_town': 'Southhampton'}, inplace=True)
```

```python
# 서로 이웃한 데이터로 대체하는 경우
df_2['deck_ffill'] = df_2['deck'].ffill()
df_2['deck_bfill'] = df_2['deck'].bfill()

df_2[['deck', 'deck_ffill', 'deck_bfill']].head(12)
```
